```python
#Importando as bibliotecas necessárias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import mean
import datetime

#Aplicando as configurações de tema
sns.set_theme(style="darkgrid", #rc=custom_params, 
              palette="colorblind", rc={'figure.figsize':(7,6)})
```

```python
#Lendo o arquivo excel
df = pd.read_excel("/content/Consumo_cerveja.xlsx")
df.head()

#Substituindo os valores para Semana e Fim de Semana
df["Final de Semana"].replace(0, "Semana", inplace= True)
df["Final de Semana"].replace(1, "Fim de Semana", inplace= True)


#Colocando Consumo de cerveja em litro
data_type_dict = {'Consumo de cerveja (litros)' : float}
df = df.astype(data_type_dict)

df.head()
```

Primeiro insight: A média de consumo de cerveja em litros é maior nos fins de semana do que nos dias de semana

```python
#Analisando a média de consumo de cerveja semana x fim de semana
sns.barplot(
    y="Consumo de cerveja (litros)", 
    x="Final de Semana",      
    data=df, 
    estimator = mean,  
    ci=None, 
    );
```

Segundo insight: Quando traçamos o temperatura média x o consumo de cerveja é possível identificar que independente da temperatura média, a tendência é que o consumo de cerveja nos fins de semana seja maior, porém, em temperaturas mais altas é possível observar maior proximidade do consumo na semana e nos fins de semana.


```python
#Analisando consumo x temperatura média x dia da semana
sns.scatterplot(data=df, x="Temperatura Media (C)", y="Consumo de cerveja (litros)",
                hue = "Final de Semana"                ,
                estimator = mean
                #hue= 
                #style = 
                #size = 
                );
```
