```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

```

```python

sns.set_theme(style="darkgrid", #rc=custom_params, 
              palette="colorblind", rc={'figure.figsize':(7,6)})


df = pd.read_excel("/content/Hikaru Yamanaka - Análise Descritiva (Medidas).xlsx")

# df.loc[df['region'] == 'INNERCITY', 'income'].sum()
df
```

### *Gráfico de barras: Salário x Região:*

```python
sns.barplot(
    x = 'income',
    y = 'region',
    data = df,
    estimator = sum,
    ci = None,
    palette = ["#414FB1","#65B3C7", "green", "#DA556D"],
    saturation = 2,
);
```

*Neste gráfico, podemos ver as regiões onde se tem os maiores salários.*

*Nela a **'INNER_CITY'** se mostra bem a frente das outras, enquanto a **'SUBURBAN'** se mostra atrás das outras regiões.*

### *Gráfico de barras: Salário x Região (Casados ou Não)*


```python
sns.barplot(
    x = 'income',
    y = 'region',
    hue = 'married',
    data = df,
    estimator = sum,
    ci = None,
    saturation = 2
);
```

*Neste gráfico, podemos ver os salários das pessoas de cada região, e se elas são casadas ou não.*

*Podemos ver neste gráfico que em 'INNER_CITY' ainda permanece como o maior salário entre 'casados' ou 'não casados'.*

*E o menor em 'SUBURBAN' com uma pequena difernça entre se são 'casados' ou 'não casados'.*

### *Gráfico de barras: Salário x Região (Com ou Sem Filhos)*

```python
sns.barplot(
    x = 'income',
    y = 'region',
    hue = 'children',
    data = df,
    estimator = sum,
    ci = None,
    saturation = 2
);
```

*Neste gráfico, podemos ver os salários das pessoas de cada região, e se elas possuem 1 ou mais filhos ou nenhum.*


---


*Podemos ver neste gráfico que em 'INNER_CITY' quem não possui filhos, possui o maior salário entre todas as regiões, já na região 'SUBURBAN' é onde nos mostra o menor salário.*


---


*No gráfico também nos mostra que, para quem tem apenas '1' filho, a região que possui o maior salário é a 'INNER_JOIN' e a região com o menor salário é a 'SUBURBAN'.*


---


*A região que possui o maior salário para pessoas que possuem '2' filhos é a 'INNER_CITY', enquanto a de menor salário permanece na 'SUBURBAN'.*


---


*E, por último, a região que tem o maior salário para pessoas que possuem '3' filhos, desta vez, foi a região 'RURAL' e a região de menor salário permanece em 'SUBURBAN'.*

### *Gráfico de Linhas: Income X Age (Sex)* 

```python
sns.lineplot(x = "age", y = "income",
             hue="sex",          
             ci = None,
             estimator = sum,
             data = df);
```

*Neste gráfico, podemos analisar as variações e picos de salários por idade e sexo.*

*Ao comparar os dois sexos, podemos ver que o 'MALE' tem uma evolução maior e mais rápida que a 'FEMALE', porém, após os 60 a 'FEMALE' permanece com o salário bem acima que a dos 'MALES' que cai drasticamente.*

### *Gráficos de Dispersão: MALE X FEMALE*

*MALE:*

```python
dfMALE = df.loc[df["sex"] == "MALE"]

sns.regplot(
    data = dfMALE, x = "age", y = "income",
);
```

Neste gráfico, podemos analisar que ele traça uma linha de tendência de 'income' e nos mostra a sua evolução conforme a sua 'age', fazendo com que a média fique mais ou menos em $ 45.000.

*FEMALE:*

```python
dfFEMALE = df.loc[df["sex"] == "FEMALE"]

sns.regplot(
    data = dfFEMALE, x = "age", y = "income",
);
```

Já neste gráfico, podemos observar que ele traça uma linha de tendência de 'income' e nos mostra a sua evolução conforme a sua 'age', fazendo com que a média fique mais ou menos em $ 41.000.
