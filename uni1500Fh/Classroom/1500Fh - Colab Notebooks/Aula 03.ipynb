```python
# ESTATÍSTICA DESCRITIVA - MEDIDAS RESUMO

import pandas as pd

df = pd.read_excel ('/content/Hikaru Yamanaka - Análise Descritiva (Medidas).xlsx')
df
```

```python
# MÉDIA

round(df['income'].mean(), 1)             

# caso queira acrescentar casa decimal colocar assim: round(df['income'].mean(), 3)
# caso não coloque nada após o mean, vai vir com muitas casas decimais
```

```python
df.loc[df['region']=='RURAL',['age']].mean()

# Na 'region' queremos os valores iguais de "RURAL", e buscamos os resultados da média de 'age'
```

```python
# MEDIANA

df['children'].median()

# Calculando a mediana
```

```python
round(df['income'].median(), 2)

# Calculando a 'income(Renda)' e utilizando round para arredondar e utilizando ', 2' para ter duas casas numéricas flutantes
```

```python
# MEDIANA

condcas = (df['married'] == 'YES')    

# condcas= condiçao casados
# Estamos utilizando a coluna 'married', para saber quem está casado 'YES' 


condfi = (df['children'] == 0)        

# condfi= condiçao filhos
# O motivo do igual '== 0' é que queremos sem filhos.


df.loc[(condcas & condfi), ['income']].median()   # mediana

# loc = reduz / condiciona
# está reduzindo as variáveis (condcas e condifi), na coluna 'income(renda)' em médiana
```

```python
# Media

condcas = (df['married'] == 'YES')    # condcas= condiçao casados
condfi = (df['children'] == 0)        # condfi= condiçao filhos

df.loc[(condcas & condfi), ['income']].mean()   # media

# está reduzindo as variáveis (condcas e condifi), na coluna 'income(renda)' em média
```
